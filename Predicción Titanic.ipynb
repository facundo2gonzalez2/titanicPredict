```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder #get_dummies

from sklearn.compose import ColumnTransformer
```

## Leo y limpio los datos

```python
titanic=pd.read_csv('titanic.csv')
```

```python
titanic
```

```python
titanic.info()
```

###### Limpio columnas:

```python
titanic=titanic.drop('Name',axis=1)
titanic=titanic.drop('PassengerId',axis=1)
titanic=titanic.drop('Cabin',axis=1)
titanic=titanic.drop('Ticket',axis=1)
titanic=titanic.drop('Embarked',axis=1)
titanic
```

```python
titanic.info()
```

```python
titanic_label=titanic['Survived']
titanic_data=titanic.drop('Survived', axis=1)
titanic_cat=titanic[['Sex']]
titanic_num=titanic_data.drop('Sex',axis=1)
```

```python
list(titanic_num)
```

```python
ohe=OneHotEncoder()
```

```python
titanic_cat_ohe=ohe.fit_transform(titanic_cat)
titanic_cat_ohe.toarray()
```

###### Aplico los pipelines:

```python
num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="mean")),#hay mas opciones aca
        ('std_scaler', StandardScaler()),
    ])

cat_pipeline=OneHotEncoder()

num_attribs = list(titanic_num)

cat_attribs = ["Sex"]

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", cat_pipeline , cat_attribs),
    ])

titanic_data_pre= full_pipeline.fit_transform(titanic_data)
```

```python
df = pd.DataFrame(titanic_data_pre)
df.info()
```

```python
df
```

## Train y test con regresión lineal

```python
x_train, x_test, t_train, t_test = train_test_split(titanic_data_pre,titanic_label, test_size=0.3, random_state=42)
```

```python
t_train
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()

modelo.fit(x_train, t_train)
```

```python
y_pred_train = modelo.predict(x_train)

print(y_pred_train[0:15])
print(t_train[0:15])
```

```python
y_pred_test = modelo.predict(x_test)
print(y_pred_test[0:15])
print(t_test[0:15])
```

```python
from sklearn.metrics import mean_squared_error

def rmse(y,t):
  return np.sqrt(mean_squared_error(y,t))
```

```python
rmse(t_train,y_pred_train)
```

```python
rmse(t_test,y_pred_test)
```
